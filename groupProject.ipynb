```python
import pandas as pd
from sklearn import preprocessing
import csv

test_df= pd.read_csv('test.csv') #read test file 
train_df= pd.read_csv('train.csv') #read test file 
store_df= pd.read_csv('store.csv') #read test file 

print("*************************test data preprocessing**********************")
print(test_df.isnull().sum()) # check if any column has null vales
test_df['Open']=test_df['Open'].fillna(1).astype(int) # fill null values in 'Open' column with 1 because null value was found for 622 store and max value for it was 1, hence assigning 1
    #convert date into number by assigning in different columns and then drop the date column
test_df['Year'] = (test_df['Date'].str.split('-').str[0]).astype(int)
test_df['Month'] = (test_df['Date'].str.split('-').str[1]).astype(int)
test_df['Day'] = (test_df['Date'].str.split('-').str[2]).astype(int)
test_df=test_df.drop(columns="Date")
    #converting "stateHoliday" column string value to integer
test_df.StateHoliday=(test_df.replace(to_replace =['a'],value=1)).astype(int)
    #remove duplicate rows from data frame
test_df=test_df.drop_duplicates()
    #scaling all the columns between 0 and 1
x = test_df.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
test_df = pd.DataFrame(x_scaled)
print("************************* Final test data ****************************")
print(test_df)

print("*************************train data preprocessing**********************")
print(train_df.isnull().sum()) # check if any column has null vales
    #convert date into number by assigning in different columns and then drop the date column
train_df['Year'] = (train_df['Date'].str.split('-').str[0]).astype(int)
train_df['Month'] = (train_df['Date'].str.split('-').str[1]).astype(int)
train_df['Day'] = (train_df['Date'].str.split('-').str[2]).astype(int)
train_df=train_df.drop(columns="Date")  
   #converting "stateHoliday" column string value to integer
train_df.StateHoliday=train_df.replace(to_replace =['a'],value =1)
train_df.StateHoliday=train_df.replace(to_replace =['b'],value =2)
train_df.StateHoliday=train_df.replace(to_replace =['c'],value =3)
train_df.StateHoliday=train_df.StateHoliday.astype(int)
    #remove duplicate rows from data frame
train_df=train_df.drop_duplicates()
    #scaling all the columns between 0 and 1
x = train_df.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
train_df = pd.DataFrame(x_scaled)
print("************************* Final train data ****************************")
print(train_df)

print("*************************store data preprocessing**********************")
print(store_df.isnull().sum()) # check if any column has null vales 
print(store_df)

   #converting "storeType" column string value to integer
store_df.StoreType=store_df.StoreType.replace(to_replace=['a','b','c','d'],value=[1,2,3,4])
   #converting "Assortment" column string value to integer
store_df.Assortment=store_df.Assortment.replace(to_replace=['a','b','c'],value=[1,2,3])
print(store_df)
   #replace the mean value of CompetitionDistance with null values in that column
rm=store_df['CompetitionDistance'].mean()
store_df['CompetitionDistance']=store_df['CompetitionDistance'].fillna(rm)
   #replace the median value of CompetitionOpenSinceMonth with null values in that column
rm=store_df['CompetitionOpenSinceMonth'].median()
store_df['CompetitionOpenSinceMonth']=store_df['CompetitionOpenSinceMonth'].fillna(rm)
   #replace the median value of CompetitionOpenSinceYear with null values in that column
rm=store_df['CompetitionOpenSinceYear'].median()
store_df['CompetitionOpenSinceYear']=store_df['CompetitionOpenSinceYear'].fillna(rm)
   #replace the median value of Promo2SinceWeek with null values in that column
rm=store_df['Promo2SinceWeek'].median()
store_df['Promo2SinceWeek']=store_df['Promo2SinceWeek'].fillna(rm)
   #replace the median value of Promo2SinceYear with null values in that column
rm=store_df['Promo2SinceYear'].median()
store_df['Promo2SinceYear']=store_df['Promo2SinceYear'].fillna(rm)
   #replace the "Jan,Apr,Jul,Oct" value of PromoInterval with null values in that column
store_df['PromoInterval']=store_df['PromoInterval'].fillna('Jan,Apr,Jul,Oct')
    #remove duplicate rows from data frame
store_df=store_df.drop_duplicates()
    #scaling all the columns between 0 and 1
x = store_df.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
store_df = pd.DataFrame(x_scaled)
print("************************* Final store data ****************************")
print(store_df)


```
